```python
import numpy
import matplotlib.pyplot as plt
from pandas import read_csv
import pandas as pd
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
import numpy as np
from keras.layers import Dropout
```

```python
data=pd.read_csv('Stock1.csv')
```

```python
data.head(5)
```

```python
#data.iloc[:,[0,6]]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))

data_scaled = scaler.fit_transform(data.iloc[:,6:7])
```

```python
features_set = []
labels = []
for i in range(40, 400):
    features_set.append(data_scaled[i-40:i, 0])
    labels.append(data_scaled[i, 0])
```

```python
data
```

```python
features_set, labels = np.array(features_set), np.array(labels)
```

```python
features_set = np.reshape(features_set, (features_set.shape[0], features_set.shape[1], 1))
```

```python
new=np.reshape(data_scaled[-40:], (data_scaled[-40:].shape[0], data_scaled[-40:].shape[1], 1))
```

```python
model = Sequential()
```

```python
model.add(LSTM(units=50, return_sequences=True, input_shape=(features_set.shape[1], 1)))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=50))
model.add(Dropout(0.2))
```

```python
model.add(Dense(units = 1))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(features_set, labels, epochs = 50, batch_size = 32)
```

```python
predictions = model.predict(features_set)
```

```python
len(labels)
```

```python
labels= np.reshape(labels, (-1, 1))
predictions_in = scaler.inverse_transform(predictions)
labels_in = scaler.inverse_transform(labels)
len(labels)
```

```python
plt.plot(predictions_in)
plt.plot(labels_in)
plt.show()
```
